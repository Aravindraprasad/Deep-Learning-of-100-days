**A perceptron is a algorithm its like a tiny brain cell in a computer that can make simple decisions. **

- Inputs: It takes in information (like numbers) from the outside world. These are called inputs. For example, if you're trying to decide if an image is of a cat, the inputs could be the colors or shapes in the image.

- Weights: Each input has a "weight," which is like importance. If an input is more important, its weight will be higher.

- Sum and Check: The perceptron adds up all the inputs after multiplying them by their weights. Then, it checks if the total is big enough (compares it to a threshold).

-  step function in a perceptron is like a simple decision-maker. It helps the    
  perceptron decide whether the output should be 1 (Yes) or 0 (No). Here's how it works:

  The perceptron calculates a value by combining inputs and their weights.
  The step function looks at this value and compares it to a threshold.
  If the value is greater than or equal to the threshold, the step function gives an output of 1 (Yes).
  If the value is less than the threshold, the step function gives an output of 0 (No).

- Output: Based on the check, it decides and gives an output: Yes (1) or No (0). For example, "Yes, this is a cat" or "No, this is not a cat."

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
df = pd. read_csv('/content/placement (1).csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_

# w1 = 40.26
# w2 = -36
```

```python
p.intercept_

# b = -25
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
